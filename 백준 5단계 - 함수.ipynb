```python
# 15596번
# 정수 n개가 주어졌을 때, n개의 합을 구하는 함수를 작성하시오.

def solve(a):
    ans = 0
    for aa in a :
        ans += aa

    return ans


    # or
# def solve(a):
#     return sum(a)
```

```python
# 4673번
# 셀프 넘버는 1949년 인도 수학자 D.R. Kaprekar가 이름 붙였다. 양의 정수 n에 대해서 d(n)을 n과 n의 각 자리수를 더하는 함수라고 정의하자. 예를 들어, d(75) = 75+7+5 = 87이다.
# 양의 정수 n이 주어졌을 때, 이 수를 시작해서 n, d(n), d(d(n)), d(d(d(n))), ...과 같은 무한 수열을 만들 수 있다. 
# 예를 들어, 33으로 시작한다면 다음 수는 33 + 3 + 3 = 39이고, 그 다음 수는 39 + 3 + 9 = 51, 다음 수는 51 + 5 + 1 = 57이다. 이런식으로 다음과 같은 수열을 만들 수 있다.
# 33, 39, 51, 57, 69, 84, 96, 111, 114, 120, 123, 129, 141, ...
# n을 d(n)의 생성자라고 한다. 위의 수열에서 33은 39의 생성자이고, 39는 51의 생성자, 51은 57의 생성자이다. 생성자가 한 개보다 많은 경우도 있다. 예를 들어, 101은 생성자가 2개(91과 100) 있다. 
# 생성자가 없는 숫자를 셀프 넘버라고 한다. 100보다 작은 셀프 넘버는 총 13개가 있다. 1, 3, 5, 7, 9, 20, 31, 42, 53, 64, 75, 86, 97
# 0000보다 작거나 같은 셀프 넘버를 한 줄에 하나씩 출력하는 프로그램을 작성하시오.


full_num = set(x for x in range(1,10001)) # 밑에서 set값 - set값을 위해서 미리 set값으로 묶어주기
num = set() # 중복 제거를 위해 set값으로 묶어주기

for i in range(1,10001) :
    for n in str(i) :
        i += int(n)
    
    num.add(i)  # list는 append, set은 add

self_num = sorted(full_num - num) # 그냥 full_num - num 만하면 정렬이 안 되어있음.

for p in self_num :
    print(p)
```

```python
# 1065번
# 어떤 양의 정수 X의 각 자리가 등차수열을 이룬다면, 그 수를 한수라고 한다. 등차수열은 연속된 두 개의 수의 차이가 일정한 수열을 말한다. 
# N이 주어졌을 때, 1보다 크거나 같고, N보다 작거나 같은 한수의 개수를 출력하는 프로그램을 작성하시오. 

num = int(input())
for_count = 0                   # 한수의 갯수 카운팅

for i in range(1, num+1) :      
    leng = str(i)               # i값을 문자열로 (인덱스 사용을 위해)
    length = len(leng)          # i가 몇글자인가
    count_list = 0              # i가 등차수열인지 아닌지 확인을 위한


    if length < 3 :             # 두 자리 숫자 이하는 무조건 등차수열을 이루기 때문에 한수가 됨

        for_count += 1          # i 값이 두 자리 이하의 숫자면 for_count에 1을 더해줌


    else:                       # 3자리 이상인 경우
        diff = ( int(leng[length-1]) - int(leng[0]) ) / int(length - 1 )

        # 첫번째자리와 마지막 빼주고 n-1자리수를 해주면 등차값이 나옴
        # 예를들어 abcd인 경우, d-a를 해주고 그걸 3으로 나눠주면, 각각 a~b값, b~c값, c~d값인 등차값이 됨


        for a in range(length-1) :
            if int(leng[a+1]) != int(leng[a]) + diff :
                count_list += 1

            # 만약 등차수열이라면 위에서 구한 등차를 이용해서
            # b값은 a에 등차를 구한 값이 될 것이고,
            # c값은 b에 등차를 구한 값이 될 것
            # 그렇기 때문에 for문과 if문을 이용해서 같지 않으면 count_list에 1을 추가해줌
            # 즉, count_list가 0이 아니라면 등차수열이 아니라는 뜻


        
        if count_list == 0 :
            for_count += 1

            # count_list가 0이라면 등차수열이기 떄문에 for_count에 1을 추가해주면서 숫자 세어주기
        
        else :
            count_list = 0
            # count_list가 0이 아닌 경우 새로운 i값을 위해 count_list값을 0으로 리셋시켜주기



print(for_count)
```
