###1978번

주어진 수 N개 중에서 소수가 몇 개인지 찾아서 출력하는 프로그램을 작성하시오.

```python
N = int(input())

number = map(int, input().split())

cnt = 0

for num in number :
    error = 0
    if num > 1 :
        for i in range(2, num) : # 2에서 n-1까지 돌 때,
            if num % i == 0 :    # 이 모든 값이 나눠떨어지게 되면 소수가 아님
                error += 1       # 자기 자신으로 나눠떨어지는건 제외해야하기 떄문에 n-1까지임
        if error == 0 :          # 그래서 error값이 없다는 건 나눠떨어진적이 없다는거고
            cnt += 1             # 그렇다는 건 소수라는 뜻

print(cnt)
                
```

###2581번

자연수 M과 N이 주어질 때 M이상 N이하의 자연수 중 소수인 것을 모두 골라 이들 소수의 합과 최솟값을 찾는 프로그램을 작성하시오.

예를 들어 M=60, N=100인 경우 60이상 100이하의 자연수 중 소수는 61, 67, 71, 73, 79, 83, 89, 97 총 8개가 있으므로, 이들 소수의 합은 620이고, 최솟값은 61이 된다.

```python
M = int(input())
N = int(input())

number = []

for num in range(M, N+1) :

    error = 0       # 소수인지 아닌지 판별을 위한 변수

    if num > 1 :        # 1은 소수도, 합성수도 아니기 때문에 1 이상인 숫자에 대해서만 생각
        for div in range(2, num) :
            if num % div == 0 :
                error += 1     
                break            # 어차피 error가 1만 올라가도 소수로 판단하니까 그냥 바로 break

        if error == 0 :
            number.append(num)

if len(number)==0 :
    print(-1)

else :
    print(sum(number))
    print(min(number))

# 1의 경우도 생각 잘 해야하니까 조심하기
```

### 1929번

### 4948번

베르트랑 공준은 임의의 자연수 n에 대하여, n보다 크고, 2n보다 작거나 같은 소수는 적어도 하나 존재한다는 내용을 담고 있다.

이 명제는 조제프 베르트랑이 1845년에 추측했고, 파프누티 체비쇼프가 1850년에 증명했다.

예를 들어, 10보다 크고, 20보다 작거나 같은 소수는 4개가 있다. (11, 13, 17, 19) 또, 14보다 크고, 28보다 작거나 같은 소수는 3개가 있다. (17,19, 23)

자연수 n이 주어졌을 때, n보다 크고, 2n보다 작거나 같은 소수의 개수를 구하는 프로그램을 작성하시오. 

```python
# 시간 초과 뜸 
# ㅇㅏ 개빡쳐

while True : 
    n = int(input())

    if n == 0 :
        break

    cnt = 0

    for a in range(n+1, 2*n+1) :
        if a == 1 :
            continue

        elif a == 2 :
            cnt += 1

        else :
            for b in range(2, int(a**0.5)+1) :   # a ** 0.5    =  >   sqrt(i) + 1
                if a % b == 0 :
                    break
            else : cnt +=1
            

    print(cnt)


```
